## Cumulative Residual RMS vs ell

```python
cl_res = hp.alm2cl(x_rec - x_true)
ell = np.arange(len(cl_res))
weights = (2 * ell + 1) / (4 * np.pi)
rms_cum = np.sqrt(
    np.cumsum(weights * cl_res)
)
fig, ax = plt.subplots(
    figsize=(10, 5),
    constrained_layout=True,
)
ax.semilogy(ell[1:], rms_cum[1:])
ax.set_xlabel(r'$\ell_{\rm max}$',
    fontsize=14)
ax.set_ylabel(
    'Cumulative residual RMS [K]',
    fontsize=14,
)
ax.set_title(
    'RMS of (recovered - true) vs '
    'included modes',
)
ax.grid(alpha=0.3)
plt.show()
```
